```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
plt.scatter(df['time_study'],df['Marks'])
plt.title('Scatter plot vs time study')
plt.xlabel('time_study')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```

```python
df_grouped=df.groupby('number_courses')['Marks'].mean().reset_index()
plt.bar(df_grouped['number_courses'],df_grouped['Marks'],color='Red')
plt.title('Number of Course vs Average mark')
plt.xlabel('number_courses')
plt.ylabel('Average marks')
plt.grid(True)
plt.show
```

```python
x=np.array(df[['time_study','number_courses']])
y=np.array(df['Marks'])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain
```

```python
ytrain
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
y_pred=model.predict(xtest)
```
